```python
import matplotlib.pyplot as plt
import numpy as np
import math

w = np.array([0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]);
beta = np.array([0, 30, 60, 90, 120, 150, 180]);

V = 10
L = 100

w_e = np.zeros([beta.size, w.size])
for i in range(0, beta.size):
    for j in range(0, w.size):
        w_e[i][j] = w[j] - ((w[j]*w[j]*V*math.cos(math.radians(beta[i])))/9.8)
    plt.plot(np.log(w), np.log(w_e[i]))
plt.show()
```

```python

WL_e = np.zeros([beta.size, w.size])
WL = np.zeros(w.size)
for i in range(0, beta.size):
    for j in range(0, w.size):
        WL_e[i][j] = 9.8*2*3.1415/(w_e[i][j]*w_e[i][j]);
        WL[j] = 9.8*2*3.1415/(w[j]*w[j]);    
    plt.plot(np.log(WL), np.log(WL_e[i]))
plt.show()
```

```python
for i in range(0, beta.size):
    for j in range(0, w.size):
        print("for angle " + str(beta[i]) + " and freq " + str(w[j]) + " V = ")
        print(9.8/(w[j]*math.cos(math.radians(beta[i]))))
```

```python
for i in range(0, beta.size):
    for j in range(0, w.size):
        print("for angle " + str(beta[i]) + " and freq " + str(w[j]) + " V = ")
        print(((w[j] - math.sqrt(9.8*2*3.1415/L))*9.8)/(w[j]*w[j]*math.cos(math.radians(beta[i]))))
```
