```python
import matplotlib.pyplot as plt
import numpy as np
import math

h = np.zeros(100)
count = -10
for i in range(0, 100):
    h[i] = count
    count = count + 0.2
h_rms = [0.5, 1, 2, 5]
y = np.zeros((4,100))
count = 0
for H_rms in h_rms:
    for i in range(0, 100):
        y[count][i] = math.exp(-2*(h[i]*h[i]/(H_rms*H_rms)))
    plt.plot(h, y[count])
    count += 1
    
plt.show()
```
